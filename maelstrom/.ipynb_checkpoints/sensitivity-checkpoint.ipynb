```python
fig,ax = plt.subplots(figsize=[4,3])

# Eclipsing binaries
x=[0.0,1.75,1.75,1.3,1.0,0.0]
y=[1.0,1.0,0.2,0.2,0.07,0.07]
ax.fill(x, y, alpha=0.25, c='orange')
ax.plot(x, y, c='orange', linestyle='dashdot')
# re '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

# Spectroscopy
x = [0.0, 3.0, 1.75, 1.2, 0.0]
y = [1.0, 1.0, 0.55, 0.35, 0.35]
ax.fill(x, y, alpha=0.25, c='red')
ax.plot(x, y, c='red', linestyle=':')

# LBI
x = [2.1,4.2,4.2,2.1]
y = [1.0,1.0,0.4,0.4]
ax.fill(x, y, alpha=0.25, c='green')
ax.plot(x, y, c='green', linestyle='dashed')
ax.plot([2.1, 2.1], [0.4, 1.], c='green', linestyle='dashed')

# PM
n = np.linspace(0, 100, 100)
x = 1.3 + n/50.0
y = 0.9747*np.exp(-1.583*x)
ax.fill_between(x, y1=y, y2=1, alpha=0.25, color='C0')
ax.plot(x, y, c='C0')
ax.plot([1.3, 1.3], [0.12, 1.], c='C0')
ax.plot([x[-1], x[-1]], [y[-1], 1], c='C0')

ax.set(xlabel=r'log$_{10}$ P$_{orb}$ [day]',
      ylabel=r'Mass ratio [$M_2 / M_1$]',
      xlim=[0,4],
      ylim=[0,1])


ax.text(1.8, 0.2, 'Pulsation timing', fontsize=9)
ax.text(0.5, 0.8, 'RV', fontsize=9)
ax.text(0.5, 0.2, 'EB', fontsize=9)
ax.text(3, 0.6, 'LBI', fontsize=9)

plt.savefig('sensitivity.pdf', dpi=300, bbox_inches='tight', pad_inches=0.)
```
