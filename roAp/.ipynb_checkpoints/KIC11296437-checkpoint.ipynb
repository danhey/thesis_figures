```python
import lightkurve as lk
import matplotlib.pyplot as plt
import numpy as np
```

```python
from astropy.timeseries import LombScargle

def amplitude_spectrum(t, y, fmin=None, fmax=None, freq=None, oversample_factor=10.0):
    """ 
    Calculates the amplitude spectrum of a given signal
    
    Parameters
    ----------
        t : `array`
            Time values 
        y : `array`
            Flux or magnitude measurements
        fmin : float (default None)
            Minimum frequency to calculate spectrum. Defaults to df
        fmax : float
            Maximum frequency to calculate spectrum. Defaults to Nyquist.
        oversample_factor : float
            Amount by which to oversample the spectrum. Defaults to 10.
    """
    tmax = t.max()
    tmin = t.min()
    df = 1.0 / (tmax - tmin)

    if fmin is None:
        fmin = df
    if fmax is None:
        fmax = 0.5 / np.median(np.diff(t))  # *nyq_mult
    if freq is None:
        freq = np.arange(fmin, fmax, df / oversample_factor)
    model = LombScargle(t, y)
    sc = model.power(freq, method="fast", normalization="psd")

    fct = np.sqrt(4.0 / len(t))
    amp = np.sqrt(sc) * fct

    return freq, amp
```

```python
lc_tess = lk.search_lightcurve('KIC 11296437', mission='TESS', exptime=120).download_all().stitch().remove_nans()
lc_kepler = lk.search_lightcurve('KIC 11296437', mission='Kepler').download_all().stitch().remove_nans()
```

```python
lc_tess.to_periodogram().plot()

lc_kepler.to_periodogram().plot()
```

```python
x_tess, y_tess = lc_tess.time.value, lc_tess.flux.value
f, a = amplitude_spectrum(x_tess, y_tess, fmin=10, fmax=12)
plt.plot(f,a)

x, y = lc_kepler.time.value, lc_kepler.flux.value
f, a = amplitude_spectrum(x, y, fmin=10, fmax=12)
plt.plot(f,a)
plt.xlim(10,12)
```

```python
x, y = lc_kepler.time.value, lc_kepler.flux.value
f, a = amplitude_spectrum(x, y, fmin=121-1, fmax=121+2)
plt.plot(f,a*1e3)
```
