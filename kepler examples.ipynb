```python
import lightkurve as lk
import matplotlib.pyplot as plt
```

```python
lc = lk.search_lightcurve('KIC 6352430', mission='Kepler', exptime=1800).download_all().stitch().remove_nans()
```

```python
x, y = lc.time.value, lc.flux.value
```

```python
plt.style.use('science')
```

```python
fig, axes = plt.subplots(3,1,figsize=[6.3, 5])

ax = axes[0]
ax.plot(x, y, '.k', rasterized=True, ms=1)
ax.set(xlim=[x[0], x[-1]])

ax = axes[1]
ax.plot(x, y, '.k', rasterized=True, ms=1)
ax.set(xlim=[825, 1000], ylabel='Normalised flux')

ax = axes[2]
ax.plot(x, y, '.k', rasterized=True, ms=1)
ax.set(xlim=[850, 860], xlabel='BJD - 2454833 [day]')

plt.savefig('precision_kepler.pdf', dpi=300, bbox_inches='tight', pad_inches=0.)
```
