```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(0.5*x)
```

```python
fig, axes = plt.subplots(3,1, figsize=[4, 5])

ax = axes[0]
ax.plot(x,np.sin(0.5*x))
ax.text(-1, 0.5, 'n=0')

ax = axes[1]
ax.plot(x,np.sin(1*x))
ax.text(-1, 0, 'n=1')

ax = axes[2]
ax.plot(x,np.sin(1.5*x))
ax.text(-1, 0, 'n=2')

[ax.axis('off') for ax in axes];
```
