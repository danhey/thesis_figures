```python
import lightkurve as lk
import matplotlib.pyplot as plt
```

```python
lc_lc = lk.search_lightcurve('KIC 7106205', mission='Kepler', exptime=1800).download_all().stitch().remove_nans()
lc_sc = lk.search_lightcurve('KIC 7106205', mission='Kepler', exptime=60).download_all().stitch().remove_nans()
```

```python
plt.style.use('science')
```

```python
fig, ax = plt.subplots(figsize=[6.3, 2])

ax.plot(lc_sc.time.value, lc_sc.flux.value, '.', c='C0')
ax.plot(lc_lc.time.value, lc_lc.flux.value, '.', c='C2')

ax.set_xlim(185, 185.5)
```
